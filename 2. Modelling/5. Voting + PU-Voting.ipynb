```python
#Нужные импорты
import pandas as pd, numpy as np
from sklearn.ensemble import VotingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from pulearn import ElkanotoPuClassifier
from sklearn.metrics import f1_score, roc_auc_score, accuracy_score, recall_score, precision_score, fbeta_score, confusion_matrix
```

```python
#Подгружаем выборку
train = pd.read_csv('train_1.csv')
test = pd.read_csv('test_1.csv')
train.head()
```

```python
test.head()
```

```python
X_train = train.drop('target', axis = 1).to_numpy()
y_train = train['target'].to_numpy()
X_test = test.drop('target', axis = 1).to_numpy()
y_test = test['target'].to_numpy()
del train, test
```

```python
F2_b = 2
mdls = ['PU_Est + Voting', 'Voting']
P = []
R = []
A = []
G = []
F1 = []
F2 = []
```

```python
#Обучаем классификаторы, которые показывали лучшие метрики
clf1 = LGBMClassifier(random_state = 42, learning_rate=0.07, num_leaves = 45, max_depth=8, min_child_samples = 55, subsample = 0.23)
clf1.fit(X_train, y_train)
clf2 = XGBClassifier(random_state = 42, n_estimators=100, learning_rate=0.15, max_depth=5)
clf2.fit(X_train, y_train)
clf3 = CatBoostClassifier(random_state = 42, n_estimators=100, learning_rate=0.15, max_depth=5)
clf3.fit(X_train, y_train)
```

```python
#Прогоняем наши классификаторы через voting и pu классификаторы
#Парамтетр soft - предсказывает метку класса на основе argmax суммы прогнозируемых вероятностей
model = VotingClassifier(estimators=[('lgbm', clf1), ('xgbc', clf2), ('cat', clf3)], voting='soft')
pu_estimator = ElkanotoPuClassifier(estimator=model, hold_out_ratio=0.1) #hold_out_ratio по дефолту
pu_estimator.fit(X_train, y_train)
```

```python
y_pred = pu_estimator.predict_proba(X_test)
Gini_w = round(2*roc_auc_score(y_test, y_pred)-1,3) #Критерий Джини

y_pred[np.where(y_pred < 0.5)[0]] = 0
y_pred[np.where(y_pred >= 0.5)[0]] = 1
```

```python
#оценки метрик
P_w = precision_score(y_test,y_pred)
A_w = accuracy_score(y_test,y_pred)
R_w = recall_score(y_test,y_pred)
F1_w = f1_score(y_test,y_pred)
F2_w = fbeta_score(y_test,y_pred,beta = F2_b)
```

```python
P.append(P_w)
R.append(R_w)
A.append(A_w)
G.append(Gini_w)
F1.append(F1_w)
F2.append(F2_w)
```

```python
#Пробуем обучить наш ансамбль без использования Pu
model = VotingClassifier(estimators=[('lgbm', clf1), ('xgbc', clf2), ('cat', clf3)], voting='soft')
model.fit(X_train, y_train)
```

```python
y_pred = model.predict_proba(X_test)[:,1]
Gini_w = round(2*roc_auc_score(y_test, y_pred)-1,3) #Критерий Джини

y_pred[np.where(y_pred < 0.5)[0]] = 0
y_pred[np.where(y_pred >= 0.5)[0]] = 1
```

```python
P_w = precision_score(y_test,y_pred)
A_w = accuracy_score(y_test,y_pred)
R_w = recall_score(y_test,y_pred)
F1_w = f1_score(y_test,y_pred)
F2_w = fbeta_score(y_test,y_pred,beta = F2_b)

P.append(P_w)
R.append(R_w)
A.append(A_w)
G.append(Gini_w)
F1.append(F1_w)
F2.append(F2_w)
```

```python
st1_res = pd.DataFrame({
        'model': mdls,
        'precision': P,
        'recall': R,
        'acc': A,
        'Gini': G,
        'F1': F1,
        'F2': F2
    })
#Метрики Voting без использования Pu выше
st1_res
```
