```python
import os
import pandas as pd
import numpy as np
import statistics
from tqdm import tqdm_notebook
from tqdm.notebook import tqdm

%matplotlib inline
```

```python
df_feat = pd.read_csv('feat_cup.csv')
print(df_feat.shape)
df_feat.sort_values(by=['dt'])
```

```python
df_objects = pd.read_csv('objects_cup.csv')
print(df_objects.shape)
print(len(df_objects['owner_id'])- len(df_objects['owner_id']. drop_duplicates ()))
df_objects.sort_values(by=['cardapp_date'])
```

```python
df_objects['cardapp_date'].astype('datetime64').sort_values().unique()
```

Удалим наблюдения с некорректными датами

```python
df_objects = df_objects.dropna(subset=['cardapp_date'])
```

```python
df_objects=df_objects[df_objects['cardapp_date'].astype('datetime64')>'2000-01-01']
```

```python
df_objects
```

Удалим подозрительные наблюдения как выбросы

```python
duplicate = dict()
for elem  in df_objects['owner_id']:
    if elem not in duplicate.keys():
        duplicate[elem] = 1
    else:
        duplicate[elem] += 1
sorted(duplicate.values(), reverse=True)[:10]
#количество повторений одинаковых owner_id(максимальные значения)
```

```python
dupl_dict = dict()
for elem in duplicate.keys():
    if  duplicate[elem] >= 20:
        dupl_dict[elem] = duplicate[elem]
        
#количество карт больше 20 вычищаем
```

```python
dupl_dict
```

Поймем, карта выпущена для себя или для другого человека

```python
df_objects['not_eq'] = np.where(df_objects['owner_id']==df_objects['holder_id'],0,1)
```

```python
df_objects['eq'] = np.where(df_objects['owner_id']==df_objects['holder_id'],1,0)
```

```python
df_objects
```

Соединим с фичами по client_id (если у client_id несколько карт в таблице, появится несколько строк с разными dt)

```python
df_agg = df_feat.merge(df_objects, left_on = 'client_id', right_on = 'owner_id', how='left')
```

```python
df_agg['dt']=df_agg['dt'].astype('datetime64')
df_agg['cardapp_date']=df_agg['cardapp_date'].astype('datetime64')
```

```python
# Временной интервал между датой открытия карты и сбором фичей
df_agg['date_diff']=(df_agg['cardapp_date']-df_agg['dt']).dt.total_seconds() / (3600*24)
```

```python
# Некорректные наблюдения (смотрим в будущее)
df_agg[df_agg['date_diff']<0].shape
```

### Формируем таргет = 1

```python
df_1 = df_agg[(df_agg['date_diff']>=0)&(df_agg['date_diff']<=31)&(df_agg['not_eq']>0)]
df_1['target']=1
```

```python
df_target_count = df_1.pivot_table(values='target',  index=['client_id','dt'], aggfunc='count').reset_index()
```

```python
df_target_count['target_count'] = df_target_count['target']
df_target_count=df_target_count.drop('target', axis=1)
```

Удалим дубликаты

```python
df_1.duplicated(['client_id', 'dt']).sum()
```

```python
df_1= df_1.drop_duplicates(['client_id', 'dt'])
```

```python
df_1=df_1.merge(df_target_count, on = ['client_id', 'dt'])
```

Удалим выбросы и некорректные значения, которые нашли вначале

```python
df_1[df_1['client_id'].isin(dupl_dict.keys())]
```

```python
df_1=df_1[~df_1['client_id'].isin(dupl_dict.keys())]
df_1['hdb_other_active_mean_psk'] = np.where(df_1['hdb_other_active_mean_psk'] <0, 0 ,df_1['hdb_other_active_mean_psk'] )
df_1['hdb_other_active_min_psk'] = np.where(df_1['hdb_other_active_min_psk'] <0, 0 ,df_1['hdb_other_active_min_psk'] )
```

```python
df_1['target_count'].max() #максимально кто-то за месяц открыл 7 карт
```

```python
df_1.to_csv('target_1.csv', index=False)
```

### Формируем таргет = 0

```python
df_0 = df_agg[(df_agg['date_diff']>=0)&(df_agg['date_diff']<=31)&(df_agg['not_eq']==0)&(df_agg['flag_target']!=1)]
```

```python
df_0=df_0.drop_duplicates(['client_id', 'dt'])
df_0['hdb_other_active_mean_psk'] = np.where(df_0['hdb_other_active_mean_psk'] <0, 0 ,df_0['hdb_other_active_mean_psk'] )
df_0['hdb_other_active_min_psk'] = np.where(df_0['hdb_other_active_min_psk'] <0, 0 ,df_0['hdb_other_active_min_psk'] )
```

```python
df_0['target']=0
```

```python
df_0.to_csv('target_0.csv', index=False)
```

### Объединяем

```python
df_1=df_1.drop(['card_uk','cardapp_date','owner_id', 'holder_id', 'not_eq', 'eq', 'date_diff', 'target_count'],axis=1)
df_0=df_0.drop(['card_uk','cardapp_date','owner_id', 'holder_id', 'not_eq', 'eq', 'date_diff'],axis=1)
```

```python
df = pd.concat([df_0,df_1])
```
